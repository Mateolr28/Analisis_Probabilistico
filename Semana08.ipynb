### Problema 1: Teorema de Chebyshev (Control de calidad)

Un fabricante de componentes electrónicos produce resistencias con un promedio de 100 ohmios y una desviación estándar de 5 ohmios. Según el Teorema de Chebyshev, ¿cuál es la probabilidad mínima de que una resistencia elegida al azar tenga un valor entre 85 y 115 ohmios? (Esto es, dentro de 3 desviaciones estándar).

Media $\mu = 100 $, Desviación estándar $\sigma = 5 $, $ k=3$ .


```python
media = 100
sigma = 5
k = 3
probabilidad_minima = 1 - (1 / k**2)
print(f"Probabilidad mínima según Chebyshev: {probabilidad_minima * 100:.2f}%")
```

### Distribución Normal (Mediciones biológicas)

Se sabe que el peso de un tipo particular de fruta sigue una distribución normal con media de 200 gramos y desviación estándar de 15 gramos. ¿Cuál es la probabilidad de que una fruta elegida al azar pese entre 185 gramos y 215 gramos?

Media $\mu = 200 $, Desviación estándar $\sigma = 15 $.


```python
from scipy.stats import norm

media = 200
sigma = 15
limite_inferior = 185
limite_superior = 215

probabilidad = norm.cdf(limite_superior, media, sigma) - norm.cdf(limite_inferior, media, sigma)
print(f"Probabilidad de que el peso esté entre 185 y 215 gramos: {probabilidad * 100:.2f}%")
```

### Distribución T-student (Estudio académico)

Un investigador desea estimar la media del rendimiento académico de un grupo de estudiantes. Toma una muestra de 12 estudiantes con un promedio de puntaje de 78 y una desviación estándar de 6. Encuentra el intervalo de confianza del 95% para la media poblacional.

$ n=12, gl=v=11, \mu = 18, \sigma = 6, \alpha=0.05 $

```python
import scipy.stats as stats
import numpy as np

n = 12
media_muestra = 78
sigma_muestra = 6
nivel_confianza = 0.95
gl = n - 1  # Grados de libertad
t_value = stats.t.ppf((1 + nivel_confianza) / 2, gl)

# Calculo del margen de error
margen_error = t_value * (sigma_muestra / np.sqrt(n))

# Intervalo de confianza
limite_inferior = media_muestra - margen_error
limite_superior = media_muestra + margen_error

print(f"Intervalo de confianza del 95%: ({limite_inferior:.2f}, {limite_superior:.2f})")
```

### Comparación Gráfica entre Distribuciones

El siguiente código genera gráficas que comparan la **Distribución Normal** y la **Distribución T-student** con diferentes grados de libertad:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
normal = stats.norm.pdf(x, 0, 1)
t_student_10 = stats.t.pdf(x, df=10)
t_student_30 = stats.t.pdf(x, df=30)

plt.figure(figsize=(10, 5))
plt.plot(x, normal, label='Distribución Normal', color='blue')
plt.plot(x, t_student_10, label='Distribución T-student (df=10)', color='red')
plt.plot(x, t_student_30, label='Distribución T-student (df=30)', color='green')
plt.title('Comparación de Distribuciones')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

1. Distribución Normal (línea azul):

    + Es simétrica y tiene una forma de campana perfectamente suave.

    + Tiene colas delgadas, lo que significa que la probabilidad de valores lejanos a la media es baja.

    + Es adecuada cuando se conoce la desviación estándar poblacional o cuando se trabaja con muestras grandes.

2. Distribución T-student con 10 grados de libertad (línea roja):

    + Tiene colas más gruesas que la distribución normal.

    + Esto ocurre porque hay más incertidumbre al estimar la media cuando se usa una muestra pequeña.

    + Al tener 10 grados de libertad, aún muestra colas más gruesas que la normal, pero no tan pronunciadas.

3. Distribución T-student con 30 grados de libertad (línea verde):

    + Se aproxima mucho a la distribución normal.

    + A medida que aumentan los grados de libertad (más datos en la muestra), la distribución T-student se vuelve más parecida a la normal.

    + Esto sucede porque a mayor tamaño de muestra, la estimación de la desviación estándar es más precisa.
